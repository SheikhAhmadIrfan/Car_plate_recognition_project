```python
import zipfile
import os

zip_path = '/content/drive/MyDrive/Cars.zip'
extract_path = '/content/drive/MyDrive/xml_files'

with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(extract_path)

# List the files in the extracted folder
os.listdir(extract_path)
```

```python
xml_folder_path = '/content/drive/MyDrive/xml_files'

for filename in os.listdir(xml_folder_path):
    if filename.endswith('.xml'):
        xml_file_path = os.path.join(xml_folder_path, filename)
        tree = xet.parse(xml_file_path)
        root = tree.getroot()

        # Process the XML data as needed
        for child in root:
            print(child.tag, child.text)
```

```python
import os
from glob import glob

xml_folder_path = glob('/content/drive/MyDrive/xml_files/Cars/*.xml')

xml_folder_path
```

```python
labels_dict=dict(filepath=[],xmin=[],xmax=[],ymin=[],ymax=[])
for filename in xml_folder_path:
        info = xet.parse(filename)
        root = info.getroot()
        member_object = root.find('object')
        labels_info = member_object.find('bndbox')
        xmin = int(labels_info.find('xmin').text)
        xmax = int(labels_info.find('xmax').text)
        ymin = int(labels_info.find('ymin').text)
        ymax = int(labels_info.find('ymax').text)
        # Print or process the extracted values
        print(f"File: {filename}, Bounding Box: ({xmin}, {xmax}, {ymin}, {ymax})")
        labels_dict['filepath'].append(filename)
        labels_dict['xmin'].append(xmin)
        labels_dict['xmax'].append(xmax)
        labels_dict['ymin'].append(ymin)
        labels_dict['ymax'].append(ymax)
```

```python
df=pd.DataFrame(labels_dict)
df
```

```python
df.to_csv('/content/drive/MyDrive/NoPlatesLabels.csv',index=False)
```
